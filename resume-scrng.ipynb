**Cleaning Data:**
1. URLs
2. hashtags
3.mentions
4.special letters
5.punctuations

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText) #self made pattern can be customised
    return cleanText
```

```python
cleanResume("my website is like this http://heloword  and access it") 
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x)) #lambda used becoz of long dataset
```

**Vactorization**
* vectorized data forms the input and output of neural networks. The training process involves manipulating and updating these vectors through layers of the network, enabling the model to learn complex relationships within the data.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
#TfidfVectorize is tool for converting text documents into numerical representations suitable for machine learning algorithms.

tfidf.fit(df['Resume'])
requiredText  = tfidf.transform(df['Resume'])
```

```python
requiredText #sparse row form
```

**Splitting**

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredText, df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

**Training & printing clasification report**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(ypred)
```
